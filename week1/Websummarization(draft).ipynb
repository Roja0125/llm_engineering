```python
## Step1 - Import statements for API calls
# imports

import requests
import ollama
from bs4 import BeautifulSoup
from IPython.display import Markdown, display
```

```python
## Step 2 - Define Model constants which includes Model URL, Header and Authorization for API call

OLLAMA_API = "http://localhost:11434/api/chat"
HEADERS = {"Content-Type": "application/json"}
MODEL = "llama3.2"

```

```python
## Step3 (1)- messages parameter is a list of dictionaries each specifying a role and content. This sets the model to maintain context across multiple turns in a conversation.

messages = [
    {"role": "system", "content": "You are a helpful assistant."},
    {"role": "user", "content": "What's the weather in Paris?"},
    {"role": "assistant", "content": "It's sunny in Paris today."},
    {"role": "user", "content": "What about tomorrow?"}
]
```

```python
## Step3 (2)- Set up payload to specify the data and parameters to guide model behavior and response. Payload is the actual data sent to the server
## Payload specifies which LLM to use, to customize responses setup temperature (parameter to control randomness), max_tokens (limits responses length), stop_sequences(defines the stoiiug oiubt fir resobse(
## Payload also provides context for the interaction

payload = {
        "model": MODEL,
        "messages": messages,
        "stream": False,
        "max_tokens": 150,
        "temperature": 0.7,
        "stop_sequences": ["\n"]
    }
```

```python
# To scrape a website use BeautifulSoup API. Some websites need you to use proper headers when fetching them:
headers = {
 "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36"
}

class Website:

    def __init__(self, url):
        """
        Create this Website object from the given url using the BeautifulSoup library
        """
        self.url = url
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.content, 'html.parser')
        self.title = soup.title.string if soup.title else "No title found"
        for irrelevant in soup.body(["script", "style", "img", "input"]):
            irrelevant.decompose()
        self.text = soup.body.get_text(separator="\n", strip=True)
```
